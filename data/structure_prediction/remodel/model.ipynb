```python
from modeller import *
from modeller.automodel import *

log.verbose()
env = environ()

# Diretório onde estão seus arquivos (PDB e .ali)
env.io.atom_files_directory = ['.']

class MyModel(automodel):
    def select_atoms(self):
        # Seleciona apenas os resíduos 122 a 165 para serem modelados (refinados)
        # O restante (166-187) ficará o mais próximo possível do template
        return selection(self.residue_range('122:A', '165:A'))

a = MyModel(env,
            alnfile  = 'alignment.ali',     # nome do arquivo de alinhamento
            knowns   = 'modelo_11',         # nome do PDB template (sem .pdb)
            sequence = 'target_model',      # nome da sequência alvo no .ali
            assess_methods=(assess.DOPE, assess.GA341))

a.starting_model= 1
a.ending_model  = 5  # Comece com poucos para testar, depois aumente para 100+
a.make()
```
