```python
import os
from collections import defaultdict

pasta = "./targets"

genes = defaultdict(set)

for f in os.listdir(pasta):
    if f.endswith("_kegg.txt"):
        gene = f.replace("_kegg.txt", "")
        genes[gene].add("kegg")
    elif f.endswith("_react.csv"):
        gene = f.replace("_react.csv", "")
        genes[gene].add("react")

ambos = [g for g in genes if len(genes[g]) == 2]
um_so = [g for g in genes if len(genes[g]) == 1]

print("Genes com os dois arquivos:", len(ambos))
print("Genes com apenas um arquivo:", um_so)
print("Total de genes únicos:", len(genes))
```
