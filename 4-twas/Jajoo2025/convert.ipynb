```python
import pandas as pd
from statsmodels.stats.multitest import multipletests

df = pd.read_excel("41467_2025_67941_MOESM8_ESM.xlsx", sheet_name=1)

mdd_cols = [c for c in df.columns if c.startswith("MDD")]
```

```python
df.shape
```

```python
df[mdd_cols].isna().sum()
```

```python
fdr = pd.DataFrame(index=df.index)

for col in mdd_cols:

    valid = df[col].notna()

    fdr[col] = float("nan")

    fdr.loc[valid, col] = multipletests(
        df.loc[valid, col],
        method="fdr_bh"
    )[1]
```

```python
fdr
```

```python
mask = (fdr < 0.05).all(axis=1)
genes = df.loc[mask, "MarkerName"]
```

```python
len(genes)
```

```python
genes_df = genes.to_frame(name="jajoo2025")

genes_df.to_csv("jajoo2025.csv", index=False)
```
