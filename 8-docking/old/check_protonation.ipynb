```python
from collections import defaultdict

FILES = {
    "sem_aguas": "PDE4B_catalytic2.pdbqt",
    "com_aguas": "fixed.pdb",
}

ionizable = {"ASP", "GLU", "HIS", "LYS", "ARG", "TYR", "CYS"}

atoms = defaultdict(lambda: defaultdict(set))

for label, filename in FILES.items():

    with open(filename) as f:
        for line in f:

            if not line.startswith(("ATOM", "HETATM")):
                continue

            resname = line[17:20].strip()

            if resname not in ionizable:
                continue

            chain = line[21].strip()
            resi = int(line[22:26])

            atom = line[12:16].strip()

            atoms[label][(chain, resi, resname)].add(atom)


def protonation(resname, atomset):

    if resname == "ASP":
        return "PROTONATED" if "HD2" in atomset or "HD1" in atomset else "DEPROTONATED"

    if resname == "GLU":
        return "PROTONATED" if "HE2" in atomset or "HE1" in atomset else "DEPROTONATED"

    if resname == "HIS":
        nd1 = "HD1" in atomset
        ne2 = "HE2" in atomset

        if nd1 and ne2:
            return "HIP"
        elif nd1:
            return "HID"
        elif ne2:
            return "HIE"
        else:
            return "UNKNOWN"

    if resname == "LYS":
        return "PROTONATED" if "HZ1" in atomset else "DEPROTONATED"

    if resname == "ARG":
        return "PROTONATED" if "HH11" in atomset else "DEPROTONATED"

    if resname == "TYR":
        return "PROTONATED" if "HH" in atomset else "DEPROTONATED"

    if resname == "CYS":
        return "PROTONATED" if "HG" in atomset else "DEPROTONATED"


all_residues = sorted(
    set(atoms["sem_aguas"]) | set(atoms["com_aguas"])
)

print("=" * 70)
print("Diferenças de protonação")
print("=" * 70)

for res in all_residues:

    state1 = protonation(res[2], atoms["sem_aguas"].get(res, set()))
    state2 = protonation(res[2], atoms["com_aguas"].get(res, set()))

    if state1 != state2:
        print(f"{res}: {state1}  -->  {state2}")
```
